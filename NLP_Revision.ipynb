```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
  sent_token=sent_tokenize(sent)
  for sentence in sent_token:
    words.append(simple_preprocess(sentence))
```

```python
def avg_word_2_vec(doc):
  valid_words = [word for word in doc if word in model.wv.index_to_key]
  if not valid_words:
    return np.zeros(model.vector_size, dtype=np.float32)
  return np.mean([model.wv[word] for word in valid_words], axis=0).astype(np.float32)
```

```python
x = []
y_filtered = [] # Create a new list for filtered labels

# Iterate through the original messages and labels
for index, row in messages.iterrows():
    message = row['messages']
    label = y[index] # Use the index to get the corresponding label from the original y

    # Preprocess the message (same preprocessing logic as before)
    review = re.sub('[^a-zA-Z]', ' ', message)
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    processed_message = ' '.join(review)

    # Tokenize the processed message for Word2Vec input
    tokenized_message = simple_preprocess(processed_message)

    # Calculate the average word vector for the tokenized message
    vector = avg_word_2_vec(tokenized_message)

    # Ensure the vector is a numpy array before appending
    if isinstance(vector, np.ndarray):
        x.append(vector)
        y_filtered.append(label)
    else:
        # Handle cases where avg_word_2_vec might return something unexpected
        print(f"Skipping message at index {index} due to invalid vector output.")


x = np.array(x)
y_filtered = np.array(y_filtered)


print(f"Length of x: {len(x)}")
print(f"Length of y_filtered: {len(y_filtered)}")

assert len(x) == len(y_filtered), "Lengths of features and labels are inconsistent."
```
